```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Suppose bank customers arrive randomly on weekday afternoons at an average of 3.2
customers every 4 minutes. What is the probability of exactly 5 customers arriving in a
4-minute interval on a weekday afternoon?

```python
y = stats.poisson(3.2) # Declare y to be a poisson random variable
x = y.pmf(5)
x = round(x,4)
print(f"The probability of 5 customers randomly arriving during a 4-minute interval {x}")
```

Bank customers arrive randomly on weekday afternoons at an average of 3.2 customers
every 4 minutes. What is the probability of having more than 7 customers in
a 4-minute interval on a weekday afternoon?

```python
x = 1 - y.cdf(7)
x = round(x,4)
print(f"probability of having more than 7 customers in a 4-minute interval {x}")
```

A bank has an average random arrival rate of 3.2 customers every 4 minutes. What
is the probability of getting exactly 10 customers during an 8-minute interval?

```python
y = stats.poisson(3.2 * 2) # Declare y to be a poisson random variable
x = y.pmf(10)
x = round(x,4)
print(f"The probability of 10 customers randomly arriving during a 8-minute interval {x}")
```

```python
print(f"Mean of y is {y.mean()}")
print(f"std of y is {y.std()}")
```

Poisson Process Simulation

```python
poisson = np.random.poisson(3.2, 10000)
poisson
```

```python
fig, ax = plt.subplots(figsize=(14,7))
ax = sns.distplot(poisson, bins=10, label='simulation results')
ax.set_xlabel("Customer Arrival 4-Minute Interval",fontsize=16)
ax.set_ylabel("Probability",fontsize=16)
```

```python
quantiles, counts = np.unique(poisson, return_counts=True)
# take the cumulative sum of the counts and divide by the sample size to
# get the cumulative probabilities between 0 and 1
cumprob = np.cumsum(counts).astype(np.double) / poisson.size
```

```python
# plot
fig, ax = plt.subplots(figsize=(14,7))
ax.plot(quantiles, cumprob,'-r',lw=4, label='CDF')
ax.set_xlabel('Arrival',fontsize=16)
ax.set_ylabel('Cumulative Probability',fontsize=16)
```

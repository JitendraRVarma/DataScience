```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import r2_score
import pickle
import warnings
warnings.filterwarnings("ignore")
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
clean_train = pd.read_csv('LGD_EDA_Train_Clean.csv')
clean_train.sample(3)
```

```python
clean_test = pd.read_csv('LGD_EDA_Test_Clean.csv')
clean_test.sample(3)
```

###  LGD Model Fitting : Recovery Rate

There is no standard zero/one inflated Beta Regression model available in pytho which can model the independent variabe constraint within [0,1]. So we will have Generalized Linear Model for modelling recovery rate

Proportion data has values that fall between zero and one. Naturally, it would be nice to have the predicted values also fall between zero and one. One way to accomplish this is to use a generalized linear model (glm) with a logit link and the binomial family. We will include the robust option in the glm model to obtain robust standard errors which will be particularly useful if we have misspecified the distribution family.

```python
X_train = clean_train.drop(['id','member_id','recovery_rate','CCF'],axis=1)
X_train  = sm.add_constant(X_train) # add intercept
y_train = clean_train['recovery_rate']
```

```python
binom_glm = sm.GLM(y_train, X_train, family=sm.families.Binomial())
binom_results = binom_glm.fit()
binom_results.summary()
```

```python
# keeping only variable for which p value is less than 0.05
important_var = ['per_pymnt','per_rec_prncp','grade_B','grade_C','grade_D','grade_E',
                 'grade_F','grade_G','initial_list_status_w']

X_train = clean_train[important_var]
X_train  = sm.add_constant(X_train) # add intercept
y_train = clean_train['recovery_rate']
```

```python
binom_glm = sm.GLM(y_train, X_train, family=sm.families.Binomial())
binom_results = binom_glm.fit()
binom_results.summary()
```

```python
y_hat_train = binom_results.predict(X_train)
y_hat_train.describe()
```

#### Test datset prediction

```python
X_test = clean_test[important_var]
X_test  = sm.add_constant(X_test) # add intercept
y_test = clean_test['recovery_rate']
```

```python
y_hat_test = binom_results.predict(X_test)
y_hat_test.describe()
```

### Model accuracy

```python
print("Train r2 score is : {}".format(r2_score(y_train, y_hat_train)))
print("Test r2 score is : {}".format(r2_score(y_test, y_hat_test)))
```

```python
test_prediction = pd.DataFrame({'Actual':y_test,'Predicted':y_hat_test})
test_prediction.sample(10,random_state=40)
```

```python
# save LGD Model
pickle.dump(binom_results,open('LGD_Model.pkl','wb'))
```
